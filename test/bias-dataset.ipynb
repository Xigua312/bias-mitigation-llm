```python
import random
import json
import os

# ==========================================
# 1. å®šä¹‰èŒä¸šå’Œæ¨¡æ¿
# ==========================================
# æ‰©å±•èŒä¸šåˆ—è¡¨ (è¦†ç›– WinoBias å¸¸è§è¯)
professions = [
    # ç”·æ€§åˆ»æ¿å°è±¡
    "The engineer", "The mechanic", "The carpenter", "The sheriff", "The CEO", 
    "The developer", "The manager", "The lawyer", "The farmer", "The driver",
    "The chief", "The guard", "The analyst", "The pilot", "The plumber",
    # å¥³æ€§åˆ»æ¿å°è±¡
    "The nurse", "The teacher", "The secretary", "The cleaner", "The receptionist",
    "The housekeeper", "The editor", "The librarian", "The cashier", "The attendant",
    "The designer", "The writer", "The assistant", "The clerk", "The tailor"
]

# æ‰©å±•å¥å¼æ¨¡æ¿ (å¤šæ ·æ€§æ˜¯å…³é”®)
templates = [
    # ç®€å•é™ˆè¿°
    "{} said that he", "{} said that she",
    "{} thought that he", "{} thought that she",
    "{} knew that he", "{} knew that she",
    "{} felt that he", "{} felt that she",
    "{} claimed that he", "{} claimed that she",
    "{} mentioned that he", "{} mentioned that she",
    
    # è¡Œä¸ºæè¿°
    "{} was working when he", "{} was working when she",
    "{} left the room because he", "{} left the room because she",
    "{} asked if he", "{} asked if she",
    
    # å½¢å®¹è¯/çŠ¶æ€
    "{} is happy because he", "{} is happy because she",
    "{} was busy so he", "{} was busy so she",
    
    # å…³ç³»
    "{} called his", "{} called her",
    "{} met his", "{} met her"
]

# ==========================================
# 2. ç”Ÿæˆæ•°æ®
# ==========================================
train_pairs = []
for prof in professions:
    # æ¯ä¸ªèŒä¸šéšæœºé‡‡æ · 10-15 ä¸ªæ¨¡æ¿ï¼Œé˜²æ­¢æ•°æ®é‡çˆ†ç‚¸ä½†ä¿è¯å¤šæ ·æ€§
    selected_templates = random.sample(templates, k=min(len(templates), 15))
    
    for temp in selected_templates:
        # æ›¿æ¢å ä½ç¬¦
        if " his" in temp or " her" in temp:
            # å¤„ç† his/her æƒ…å†µ
            base = temp.replace(" his", "").replace(" her", "")
            pair = (base.format(prof) + " his", base.format(prof) + " her")
        else:
            # å¤„ç† he/she æƒ…å†µ
            base = temp.replace(" he", "").replace(" she", "")
            pair = (base.format(prof) + " he", base.format(prof) + " she")
            
        train_pairs.append(pair)

# æ‰“ä¹±æ•°æ®
random.shuffle(train_pairs)

print(f"âœ… æ•°æ®ç”Ÿæˆå®Œæˆï¼")
print(f"   èŒä¸šæ•°é‡: {len(professions)}")
print(f"   æ ·æœ¬æ€»æ•°: {len(train_pairs)} å¯¹")

# ==========================================
# 3. ä¿å­˜åˆ° dataset æ–‡ä»¶å¤¹
# ==========================================
# å®šä¹‰ä¿å­˜ç›®å½•å’Œæ–‡ä»¶å
output_dir = "dataset"
file_name = "gender_counterfactual_train.json" # æ˜ç¡®æ ‡æ³¨æ˜¯ Gender æ•°æ®
output_path = os.path.join(output_dir, file_name)

# ç¡®ä¿ dataset ç›®å½•å­˜åœ¨
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    print(f"ğŸ“‚ åˆ›å»ºç›®å½•: {output_dir}")

# å†™å…¥ JSON æ–‡ä»¶
with open(output_path, "w", encoding="utf-8") as f:
    json.dump(train_pairs, f, indent=4)

print(f"ğŸ’¾ æ•°æ®é›†å·²æˆåŠŸä¿å­˜è‡³: {os.path.abspath(output_path)}")
print(f"ğŸ“ æ–‡ä»¶è¯´æ˜: è¯¥æ–‡ä»¶åŒ…å« {len(train_pairs)} å¯¹ç”¨äºæ€§åˆ«å»åï¼ˆGender Debiasingï¼‰çš„åäº‹å®å¥å­ã€‚")
```
