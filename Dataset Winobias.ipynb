```python
import json
from pathlib import Path

DATA_DIR = Path("dataset/Winobias")

FILES = {
    "pro": DATA_DIR / "pro_stereotyped_type1.txt.test",
    "anti": DATA_DIR / "anti_stereotyped_type1.txt.test"
}

OUTPUT = DATA_DIR / "winobias_type1_test.jsonl"

def parse_line(line):
    """
    Winobias Type1 lines have a fixed structure.
    Example:
    'The doctor hired the nurse because he was short on staff.'
    """
    words = line.strip().split()
    A = words[1]   # e.g. doctor
    B = words[4]   # e.g. nurse
    return {
        "sentence": line.strip(),
        "A": A,
        "B": B
    }

count = 0
with open(OUTPUT, "w") as fout:
    for split, path in FILES.items():
        with open(path) as f:
            for line in f:
                if not line.strip():
                    continue
                item = parse_line(line)
                item["type"] = split  # pro / anti
                fout.write(json.dumps(item) + "\n")
                count += 1

print(f"Saved {count} examples to {OUTPUT}")
```
